```
class Stack :
  def __init__( self ):
    self.__datos = []

  def is_empty( self ):
    return len(self.__datos) == 0

  def get_top( self ):
    return self.__datos[-1]

  def pop( self ):
    return self.__datos.pop()

  def push( self , elemento ):
    self.__datos.append( elemento )

  def get_length( self ):
    return len(self.__datos)

  def to_string( self ):
    print("----------")
    for j in self.__datos[-1::-1]:
      print(f" { j } ")
    print("----------")


def main():
  
	#Lectura de bloc de notas
	fic = open("codigo1.txt" , "r")
	lineas_Codigo_C = fic.readlines()
	fic.close()
	p = Stack()
	
	print("El codigo a evaluar es el siguiente:\n\n")

	#Imprimimos codigo de bloc de notas
	for i in range(len(lineas_Codigo_C)):
		print(lineas_Codigo_C[i])

	#Evaluamos linea por linea para ver si tiene "( )" o "{ }" o "[ ]"
	for i in range(len(lineas_Codigo_C)):
		for j in lineas_Codigo_C[i]:
			if j == "(" or j == "{" or j == "[":
				p.push("@")
			elif j == ")" or j == "}" or j == "]":
				p.pop()
			
	if p.is_empty():
		print("---------------------------")
		print("El codigo está balanceado")
		print("---------------------------")
	else:
		print("---------------------------")
		print("El codigo no está balanceado")
		print("---------------------------")

  #CODIGO 2
	#Lectura de bloc de notas
	fic = open("codigo2.txt" , "r")
	lineas_Codigo_C = fic.readlines()
	fic.close()
	p = Stack()
	
	print("El codigo a evaluar es el siguiente:\n\n")

	#Imprimimos codigo de bloc de notas
	for i in range(len(lineas_Codigo_C)):
		print(lineas_Codigo_C[i])

	#Evaluamos linea por linea para ver si tiene "( )" o "{ }" o "[ ]"
	for i in range(len(lineas_Codigo_C)):
		for j in lineas_Codigo_C[i]:
			if j == "(" or j == "{" or j == "[":
				p.push("@")
			elif j == ")" or j == "}" or j == "]":
				p.pop()
			
	if p.is_empty():
		print("---------------------------")
		print("El codigo está balanceado")
		print("---------------------------")
	else:
		print("---------------------------")
		print("El codigo no está balanceado")
		print("---------------------------")

main()
```
